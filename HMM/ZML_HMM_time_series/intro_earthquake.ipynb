# Earthquake time series

**Data:**

Series of annual counts of major earth- quakes (i.e. magnitude 7 and above) for the years 1900–2006

```r
eartquake_data <- read.table("./data/earthquakes.txt")
```

```r
t <- eartquake_data[,1]
y <- eartquake_data[,2]
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
plot(eartquake_data, xlab='Year', ylab='Earthquake count', type='o', cex=0.3)
```

**Notes:**

- the usual model for unbounded counts is the Poisson distribution
- not appropriate: autoregressive moving-average (ARMA), as such models are based on the normal distribution
- we will see that the series displays considerable overdispersion relative to the Poisson distribution, and strong positive serial dependence

Advantage of using an HMM:
- allows the probability distribution of each observation to depend on the unobserved (or ‘hidden’) state of a Markov chain
- can accommodate both overdispersion and serial dependence

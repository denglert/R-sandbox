\begin{equation}
    Cov(X,Y) = E\left[  (X - E[X]) (Y-E[Y])  \right]
\end{equation}

\begin{equation}
    Cor(X,Y)
    =
    \frac{Cov(X,Y)}{\sigma(X), \sigma(Y)}
\end{equation}

$Cov(DAX,FTSE)$

```r
cov(DAX_logreturns, FTSE_logreturns)
```

$Cor(DAX,FTSE)$

```r
cor(DAX_logreturns, FTSE_logreturns)
```

## Time series objects


- `ts()`
- `is.ts()`


```r
?ts
```

Example:
Brownian motion

```r
t <- 0:100  # time
sig2 <- 0.01
```

```r
dx <- rnorm(n = length(t) - 1, sd = sqrt(sig2))
x <- c(0, cumsum(dx))
```

```r
plot(x)
```

```r
class(x)
```

```r
typeof(x)
```

```r
is.ts(x)
```

Create a `time-series` object from `x`

```r
time_series <- ts(x, start=2010, frequency=2)
```

```r
plot(time_series)
```

```r
typeof(time_series)
```

```r
class(time_series)
```

```r
is.ts(time_series)
```

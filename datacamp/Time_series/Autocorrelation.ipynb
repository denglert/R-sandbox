# Autocorrelation

```r
eu_stocks <- EuStockMarkets
```

```r
ftse <- eu_stocks[,4]
```

```r
nrows <- length(ftse)
```

```r
head(ftse,n=3)
```

```r
tail(ftse,n=3)
```

Drop the last value

```r
tail(ftse[-nrows],n=3)
```

Drop the first value

```r
head(ftse[-1],n=3)
```

Drop the first two elements

```r
ftse[-(1:2)]
```

Autocorrelation between the consecutive values

```r
cor(ftse[1:(nrows-1)],ftse[2:nrows])
```

```r
cor(ftse[-nrows],ftse[-1])
```

```r
deltat(ftse)
```

```r
ftse_logreturn <- diff(log(ftse))
```

```r
acf(ftse_logreturn, lag.max=2, plot=FALSE)
```

```r
acf(ftse_logreturn)
```

```r
# Define x_t0 as x[-1]
x_t0 <- ftse_logreturn[-1]

# Define x_t1 as x[-n]
x_t1 <- ftse_logreturn[-length(ftse_logreturn)]
```

```r
n <- length(ftse_logreturn)
```

```r
# Confirm that x_t0 and x_t1 are (x[t], x[t-1]) pairs  
head(cbind(x_t0, x_t1))
```

```r
plot(x_t0, x_t1)
```

```r
cor(x_t0, x_t1)
```

```r
acf(ftse_logreturn, lag.max = 1, plot = FALSE)
```

```r
cor(x_t1, x_t0) * (n-1)/n
```

```r
x <- seq(from = -5, to = 5, by = 0.01)
```

```r
T = 1.0
```

```r
y <- sin(x*2.0*pi/T)
```

```r
z <- y+rnorm(length(y), sd=0.5)
```

```r
par(mfrow=c(2,1))
plot(x,y)
plot(x,z)
```

```r
par(mfrow=c(2,1))
acf(y)
acf(z)
```

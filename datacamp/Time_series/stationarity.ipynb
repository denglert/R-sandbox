# Stationarity

Weak stationarity holds if the following model parameters are constant and
- mean, $E[X_{t}]$
- variance, $[X_{t}]$
- covariance, $C_{s} = Cov[X_{t}, X_{t-s}]$ only depends on $s$

```r
white_noise <- arima.sim(model=list(order=c(0,0,0)), n=100)
ts.plot(white_noise)
```

```r
random_walk <- cumsum(white_noise)
ts.plot(random_walk)
```

```r
wn_drift <- arima.sim(model=list(order = c(0, 0, 0)), n=100, mean=0.4)
ts.plot(wn_drift)
```

```r
rw_drift <- cumsum(wn_drift)
ts.plot(rw_drift)
```

```r
plot.ts(cbind(white_noise, random_walk, wn_drift, rw_drift))
```

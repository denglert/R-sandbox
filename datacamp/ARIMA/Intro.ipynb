## Time series regression models

**Regression:**

\begin{equation}
    Y_{i}
    =
    \beta X_{i} + \epsilon_{i}
\end{equation}

where $\epsilon_{i}$ is white noise.


**Autoregression:**
\begin{equation}
    X_{t}
    =
    \Phi X_{t-1} + \epsilon_{i}
\end{equation}

where $\epsilon_{i}$ is white noise.

**Moving average:**

\begin{equation}
    \epsilon_{t}
    =
    W_{t} + \theta W_{t-1}
\end{equation}

where $W_{t}$ is white noise.

**ARMA model:**
\begin{equation}
    X_{t}
    =
    \phi X_{t-1}
    +
    W_{t}
    +
    \theta W_{t-1}
\end{equation}

White noise:
- independent normals with common variance (homoscedastic)
- basic building block of time series
